#### 15. Crie uma matriz 2D com 1 na borda e 0 no interior (★☆☆)

````python
def borda_1(linhas, colunas):
    matriz = np.ones((linhas, colunas), dtype=int)
    matriz[1:-1,1:-1] = 0
    return matriz

print(borda_1(4,6),'\n')
print(borda_1(3,3),'\n')
print(borda_1(7,10))
````

#### 16. Como adicionar uma borda (preenchida com zeros) em torno de um array existente? (★☆☆)

````python
print(matriz_10_10, '\n')

def borda_0(matriz_ref, c_v=0):
    Z = np.pad(matriz_ref, pad_width=1, mode='constant', constant_values=c_v)
    return Z

print(borda_0(matriz_10_10))
````

#### 17. Qual é o resultado da seguinte expressão? (★☆☆)
```python

```

````python
print('0 * np.nan:', 'nan')
print('np.nan == np.nan:', 'False')
print('np.inf > np.nan:', 'False')
print('np.nan - np.nan:', 'nan')
print('np.nan in set([np.nan]):', 'True')
print('0.3 == 3 * 0.1:', 'False')
````
